```python
## Phân tích tâp dữ liệu Friday
# importing pandas package 
import pandas as pd 
import time
seconds = time.time()


# making data frame from csv file  
data = pd.read_csv("./CSVs/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv") 
  
# generating one row  
rows = data.sample(frac =0.1) 
  
# checking if sample is 0.25 times data or not 
  
if (0.1*(len(data))== len(rows)): 
    print( "Cool") 
    print(len(data), len(rows)) 
##display 
##rows
export_csv = rows.to_csv (r'./CSVs/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX_new.csv', index = None, header=True) #Don't forget to add '.csv' at the end of the path
print (rows)
print("mission accomplished!")
print("Total operation time: = ",time.time()- seconds ,"seconds")

```
